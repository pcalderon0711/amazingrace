```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
nums = list(map(str,range(1,30)))
link = 'https://en.wikipedia.org/wiki/The_Amazing_Race_'
```

```python
contestants = []
for num in nums:
    print(num)
    dat = requests.get(link+num)
    soup = BeautifulSoup(dat.content, 'lxml')
    table = soup.select('table.wikitable')[0]
    
    num_dat = len(table.select('tr')[2].select('td'))
    if (int(num) < 25) or (int(num) > 27):
        num_legs = num_dat-2
        skip = 1
    else:
        num_legs = num_dat-3
        skip = 2

    for i in range(2,len(table.select('tr'))):
        name = table.select('tr')[i].th.text
        tds = table.select('tr')[i].select('td')
        contestant = {'name':name, 'rank':i-1, 'season':num}
        for j in range(len(tds)-1):
            if j==0:
                if int(num) != 29:
                    contestant['relation'] = tds[j].text
                else:
                    contestant['relation'] = 'Strangers'
            else:
                if tds[j].text == '' or (((int(num) > 24) and (int(num) < 28)) and j == 1):
                    continue
                contestant[j-skip+(13-num_legs)] = int(re.sub('[^0-9]+[0-9]*','', tds[j].text))

        contestants.append(contestant)
```

```python
def get_rolling_mean(series):
    series = series.dropna()
    out = []
    for i in range(len(series)):
        out.append(series[:i+1].mean())
    return pd.Series(out, index=series.index)
```

```python
def return_rolling_df(dd):
    return dd[ncols].apply(get_rolling_mean,axis=1)#.fillna(method='ffill',axis=1)
```

```python
roll_df = return_rolling_df(scaled)
roll_df['name'] = scaled['name']
roll_df['season'] = scaled['season']
roll_df['rank'] = scaled['rank']
roll_df = roll_df[['name','season','rank',*ncols]]
```
